# 对于上证指数成分股的简单多因子筛选策略

多因子模型是量化投资领域应用最广泛也是最成熟的量化选股模型之一，建立在投资组合、资本资产定价（CAPM）、套利定价理论（APT）等现代金融投资理论基础上。多因子模型假设市场是无效或弱有效的，通过主动投资组合管理来获取超额收益。多因子选股的核心思想在于，市场影响因素是多重的并且是动态的，但是总会有一些因子在一定的时期内能发挥稳定的作用。在量化实践中，由于不同市场参与者或分析师对于市场的动态、因子的理解存在较大差异，一般采用一系列的因子作为选股标准，满足则买入，不满足则卖出。不同的市场时期总有一些因子在发挥作用，该模型相对来说比较稳定。模型的优点是可以综合很多信息后给出一个选股结果。选取的因子不同以及如何综合各个因子得到最终判断的方法不同会产生不同的模型。一般来说，综合因子的方法有打分法和回归法两种，打分法较为常见。本代码先对因子进行回归法分析，然后人工选择几个具有显著性并且系数较大的因子，对这几个因子根据系数加权平均求出每个股票的rank，rank并不是越高越好，只是代表了一种排名，要根据具体情况来分析。由于采取数据来自于聚宽，对于数据的采集有上限，本代码只从2019-01-01购买入手，观察至2019年12月31日，选用一年的数据进行分析，对于2020年，由于部分数据缺失不做评价，若要进行分析，只需要改动下面“xxxx-xx-xx”字样的时间即可，并且根据结果来进行选股的策略，当需要改动因子的时，只需要改动factors，改动成分股或者股票池的时候，只需要改变stock_set

### 获取数据、回归和数据清理

由于聚宽的因子库需要收费，因此先计算、收集比较简单和常用的指标，所以结合JQ能提供的数据，具体选取以下三个方面的因子：

（1）估值：账面市值比（B/M)、盈利收益率（EPS）、动态市盈（PEG）

（2）成长性：ROE、ROA、主营毛利率（GP/R)、净利率(P/R)

（3）资本结构：资产负债（L/A)、固定资产比例（FAP）、流通市值（CMV）

```python
factors  = ['B/M','EPS','PEG','ROE','ROA','GP/R','P/R','L/A','FAP','CMV']

#取出因子数值
def get_factors(fdate,factors):
    stock_set = get_index_stocks('000001.XSHG',fdate)
    q = query(
        #income.statDate,
        valuation.code,
        balance.total_owner_equities/valuation.market_cap/100000000,
        income.basic_eps,
        valuation.pe_ratio,
        income.net_profit/balance.total_owner_equities,
        income.net_profit/balance.total_assets,
        income.total_profit/income.operating_revenue,
        income.net_profit/income.operating_revenue,
        balance.total_liability/balance.total_assets,
        balance.fixed_assets/balance.total_assets,
        valuation.circulating_market_cap
        ).filter(
        valuation.code.in_(stock_set),
        valuation.circulating_market_cap
    )
    fdf = get_fundamentals(q, date=fdate)
    fdf.index = fdf['code']
    fdf.columns = ['code'] + factors
    return fdf.iloc[:,-10:]
fdf = get_factors('2019-01-01',factors)
fdf
#fdf.to_excel('/Users/lezhao/Desktop/ss.xlsx')  #为了查看数据，到excel观察，观察后在后续的回归分析中舍弃掉了GP/R，P/R
```

```python
def get_prices(codes_list):
    prices = []
    for cd in codes_list:
        df = get_price(cd, count = 1, end_date='2019-03-31', frequency='daily', fields=['close'])
        prices.append(df['close'][0])
    return prices

prices_list = get_prices(codes_list)
prices_list
```

回归

```python
X = fdf[['B/M','EPS','PEG','ROE','ROA','L/A','FAP','CMV']]
est = sm.OLS(prices_list,X).fit()
est.summary()
```

选取绝对值比较大的系数的三个指标EPS、ROE、ROA并且根据其系数定义其权重

根据聚宽提供的接口：
get_price()在获得一个股票的时候会生成index为时间的dataframe。
但在获得多个股票和多个日期的时候，会自动生成一个panel，index比较混乱。
但是这样不利于分析，下面的merge_all()为自己定义的函数接口，在内部使用了get_price()获得单个股票从而获得index为时间的dataframe，并且传入的part为一个代码的list，对于这个list的每一个代码进行get_price()，将获得的数据拼接在一起生成新的数据集。

```python
def merge_all(pre,part):
    for i in range(1,len(part)):
        tmp1 =get_price(security=part[i],start_date='2019-01-01',end_date='2020-01-01',frequency='daily',panel=False)
        tmp2 = DataFrame(tmp1['close'])
        tmp2.columns = [part[i]]
        pre = pd.merge(pre,tmp2,left_index=True,right_index=True)
    return pre
```

```python
pre11 = get_price(security=part1[0],start_date='2019-01-01',end_date='2020-01-01',frequency='daily',panel=False)
pre12 = DataFrame(pre11['close'])
pre12.columns = [part1[0]]
group1 = merge_all( pre12 , part1)
```
